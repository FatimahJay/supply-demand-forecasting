```python
import os
import pandas as pd

folder_path = "/content/drive/MyDrive/AI-Project Dataset/training_data/cluster_map"
files = os.listdir(folder_path)

# Create an empty list to store all the file contents
data = []

# Loop through all the files in the folder and append their contents to the list
for file in files:
    file_path = os.path.join(folder_path, file)
    with open(file_path, "r") as f:
        file_content = f.read()
    # Split the file content into rows and append to the data list
    data.extend(file_content.strip().split('\n'))

# Split each row into columns and create a DataFrame
df_cluster = pd.DataFrame([x.split('\t') for x in data], columns=['region_hash', 'region_id'])

print(df_cluster)
```

```python
import os

folder_path = "/content/drive/MyDrive/AI-Project Dataset/training_data/order_data"
files = os.listdir(folder_path)

# Create an empty list to store all the file contents
data = []

# Loop through all the files in the folder and append their contents to the list
for file in files:
    file_path = os.path.join(folder_path, file)
    with open(file_path, "r") as f:
        file_content = f.read()
    # Split the file content into rows and append to the data list
    data.extend(file_content.strip().split('\n'))

# Split each row into columns and create a DataFrame
df_order = pd.DataFrame([x.split('\t') for x in data], columns=['order_id', 'driver_id', 'passenger_id', 'start_region_hash','dest_region_hash','price','time'])

# Write the DataFrame to a CSV file
print(df_order.head())
```

```python
df_order=df_order.drop(['order_id','passenger_id'],axis=1)
```

```python
print(df_order.head())
```

```python
df_order['start_region_hash']=df_order['start_region_hash'].map(df_cluster.set_index('region_hash')['region_id'])
```

```python
df_order.head()
```

```python
df_order['dest_region_hash']=df_order['dest_region_hash'].map(df_cluster.set_index('region_hash')['region_id'])
```

```python
df_order.head()
```

```python
df_order['driver_id']=df_order['driver_id'].replace('NULL', 0)
df_order.head()
```

```python
def null_values_Replacement(val):
    if val==0:
        return 0
    else:
        return 1
```

```python
df_order['driver_id']=df_order['driver_id'].apply(null_values_Replacement)
df_order.head()
```

```python
import pandas as pd

# read the order table
order_df = pd.DataFrame()
order_df = df_order

# convert the 'time' column to datetime format
order_df['time'] = pd.to_datetime(order_df['time'])

# create 144 time slots
bin_range = pd.date_range(start=order_df['time'].min(), end=order_df['time'].max(), freq='10T')
order_df['time_slot'] = pd.cut(order_df['time'], bins=bin_range, labels=False)

# print the updated order table
order_df.head(501286)
```

```python
#read the weather table
weather_df = pd.DataFrame()
weather_df = df_weather

# Convert the 'Time' column to datetime format
weather_df['Time'] = pd.to_datetime(weather_df['Time'])

# Create 144 time slots
bin_range = pd.date_range(start=weather_df['Time'].min(), end=weather_df['Time'].max(), freq='10T')
weather_df['time_slot'] = pd.cut(weather_df['Time'], bins=bin_range, labels=False)

weather_df.head()
```

```python
weather_grouped = weather_df.groupby(['time_slot'])
weather_grouped.head()
```

```python
# Merge the two data frames on the 'time_slot' column
# Group the order data frame by 'time_slot'
grouped_order_df = order_df.groupby('time_slot').first().reset_index()
merged_df = pd.merge(grouped_order_df, weather_df, on='time_slot')

# Print the final merged data frame
print(merged_df)
```

```python
merged_df['demand']=1
merged_df.head()
```

```python
merged_df = merged_df.drop(['time', 'Time','dest_region_hash'], axis=1)
merged_df.head()
```

```python
merged_df = merged_df.drop(['PM2.5'], axis=1)
merged_df.head()
```

```python
merged_df['price'] = pd.to_numeric(merged_df['price'])
final_df = merged_df.groupby(['start_region_hash', 'time_slot']).agg({'price':'sum','driver_id':'sum','demand':'count','temperature': lambda x: pd.to_numeric(x).mean(),'Weather': lambda x: x.mode()[0]}).reset_index()
print(final_df)
```

```python
final_df.columns=['region','time','price','supply','demand','temperature','weather']
print(final_df)
```

```python
final_df['gap']=final_df['demand']-final_df['supply']
print(final_df)
```

```python
from sklearn.linear_model import LinearRegression 
linear_algo=LinearRegression()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_df.drop(['gap'],axis=1), final_df['gap'], test_size=0.25, random_state=101)
X_train
```

```python
linear_algo.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = linear_algo.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```

```python
import numpy as np
predict=np.int_(linear_algo.predict(X_test))
from sklearn.metrics import classification_report
print(classification_report(predict,y_test))
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, predict)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values for Training Set')
plt.show()
```
